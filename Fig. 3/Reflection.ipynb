```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import h5py
```

```python

# Load MATLAB data file
mat_file = h5py.File('COMBINED.mat')


# Extract data from file and assign to variables
f = mat_file['f'][:]
r1 = mat_file['R_bare'][:]
r2 = mat_file['R_st'][:]

lightspeed = 299792458  # m/s
wavelength = (lightspeed*1e9)/f

# Transpose the array
w = wavelength.T
R1 = r1.T
R2 = r2.T

#--------------------------------#

plt.rcParams['font.family'] = 'Times New Roman'

fig = plt.figure()
fig,axs = plt.subplots(figsize=(5,3.5))

line1=axs.plot(w,R1,color=('red'),linewidth=1.5,linestyle='-',label='Bare CdTe')
line2=axs.plot(w,R2,color=('green'),linewidth=1.5,linestyle='-',label='CdTe with NC texture')

# axs.set_title('Reflection vs Wavelength',fontsize=14,color=('black'))
axs.axis([400,1100,     -0.01,1])
axs.set_xlabel('Wavelength (nm)',fontsize=16,color=('black'))
axs.set_ylabel('Reflection',fontsize=16,color=('black'))

plt.legend(fontsize=12)

# Customize tick marks (inside, bigger size, and thicker)
axs.tick_params(direction='in', which='both', labelsize=13)
axs.minorticks_on()

plt.savefig('reflection.pdf', format='pdf', dpi=600, bbox_inches='tight')

fig.set_dpi(500)
```
