```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import h5py
```

```python
# Load MATLAB data file
mat_file = h5py.File('enhancement_factor.mat')

# Extract data from file and assign to variables
f = mat_file['f'][:]

af1 = mat_file['A_bare'][:]
af2 = mat_file['A_st'][:]
af3 = mat_file['A_stnp'][:]


lightspeed = 299792458  # m/s
wavelength = (lightspeed*1e9)/f
wavelength_point = 701

# Transpose the array
w = wavelength.T
AF1 = af1.T
AF2 = af2.T
AF3 = af3.T

enhancement_factor_bare = AF1/AF1
enhancement_factor1 = AF2/AF1
enhancement_factor2 = AF3/AF1

enh_sum1 = np.sum(enhancement_factor1)
enh_sum1 = enh_sum1/wavelength_point

enh_sum2 = np.sum(enhancement_factor2)
enh_sum2 = enh_sum2/wavelength_point

#--------------------------------#

plt.rcParams['font.family'] = 'Times New Roman'

fig = plt.figure()
fig,axs = plt.subplots(figsize=(5,2.7))
# axs.grid(True)

line1=axs.plot(w,enhancement_factor_bare,color=('grey'),linewidth=1,linestyle='-',label='Bare')
line2=axs.plot(w,enhancement_factor1,color=('red'),linewidth=1,linestyle='--',label='Nanocone texture')
line3=axs.plot(w,enhancement_factor2,color=('blue'),linewidth=1,linestyle='-.',label='Nanocone texture & Ge nanoparticle')

axs.set_xlabel('Wavelength (nm)',fontsize=14,color=('black'))
axs.set_ylabel('Enhancment factor',fontsize=14,color=('black'))

# Add the legend with transparent background
legend = axs.legend(fontsize=9)
legend.get_frame().set_alpha(0.3)  # Set the frame alpha to 0 for transparency

# Add shaded regions
axs.axvspan(400, 800, facecolor='green', alpha=0.15)
axs.axvspan(800, 1100, facecolor='orange', alpha=0.15)

print(f'Enhancement factor value for Texture: {enh_sum1:.4f}')
print(f'Enhancement factor value for Texture & Nanoparticle: {enh_sum2:.4f}')

# Customize tick marks (inside, bigger size, and thicker)
axs.tick_params(direction='in', which='both', labelsize=11)
axs.minorticks_on()

plt.savefig('enh_factor.pdf', format='pdf', dpi=600, bbox_inches='tight')

fig.set_dpi(500)
```
