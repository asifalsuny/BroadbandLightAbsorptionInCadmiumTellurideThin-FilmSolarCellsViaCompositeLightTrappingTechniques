```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import h5py

# Load MATLAB data file
mat_file = h5py.File('COMBINED.mat')

# Extract data from file and assign to variables
f = mat_file['f'][:]
a1 = mat_file['A_bare'][:]
a2 = mat_file['A_st'][:]
a3 = mat_file['A_stnp'][:]

lightspeed = 299792458  # m/s
wavelength = (lightspeed*1e9)/f

# Transpose the array
w = wavelength.T
A1 = a1.T
A2 = a2.T
A3 = a3.T

#--------------------------------#
plt.rcParams['font.family'] = 'Times New Roman'

fig = plt.figure()
fig,axs = plt.subplots(figsize=(5,2.7))
# axs.grid(True)

line1=axs.plot(w,A1,color=('grey'),linewidth=1,linestyle='-',label='Bare CdTe')
line2=axs.plot(w,A2,color=('red'),linewidth=1,linestyle='-.',label='CdTe with NC texture')
line3=axs.plot(w,A3,color=('blue'),linewidth=1,linestyle='--',label='CdTe with NC texture & Ge NP')

axs.set_xlabel('Wavelength (nm)',fontsize=14,color=('black'))
axs.set_ylabel('Absorption',fontsize=14,color=('black'))

# Add shaded regions
axs.axvspan(400, 800, facecolor='green', alpha=0.15)
axs.axvspan(800, 1100, facecolor='orange', alpha=0.15)

# Add the legend with transparent background
legend = axs.legend(fontsize=9)
legend.get_frame().set_alpha(0.3)  # Set the frame alpha to 0 for transparency

# Customize tick marks (inside, bigger size, and thicker)
axs.tick_params(direction='in', which='both', labelsize=11)
axs.minorticks_on()

plt.savefig('absorption.pdf', format='pdf', dpi=600, bbox_inches='tight')

fig.set_dpi(500)
```
