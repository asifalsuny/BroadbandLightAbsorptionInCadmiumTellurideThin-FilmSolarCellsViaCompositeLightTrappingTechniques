```python
import h5py
import matplotlib.pyplot as plt

# Load MATLAB data file
mat_file = h5py.File('T&R_orders_vs_wavelength.mat')

# Extract data from file and assign to variables
wavelength = mat_file['wavelength'][:]
a1 = mat_file['T_orders'][:]
a2 = mat_file['R_orders'][:]

# Transpose the array
w = wavelength.T
A1 = a1.T
A2 = a2.T

#--------------------------------#

plt.rcParams['font.family'] = 'Times New Roman'


fig, axs = plt.subplots(figsize=(4.5, 3))
# fig, axs = plt.subplots()

line1 = axs.plot(w, A1, color='red', linewidth=1.5, linestyle='-', label='Transmission')
line2 = axs.plot(w, A2, color='orange', linewidth=1.5, linestyle='-', label='Reflection')

axs.set_xlabel('Wavelength (nm)', fontsize=14, color='black')
axs.set_ylabel('Number of supported orders', fontsize=14, color='black')

# Set legend font size
plt.legend(fontsize=10)

# Adjust y-axis limits
# axs.set_ylim([-0.5, 4])  # Set the y-axis limits as desired

# Customize tick marks (inside, bigger size, and thicker)
# axs.tick_params(direction='in', length=6, width=1.5, which='both', axis='x')
axs.tick_params(direction='in', which='both')
axs.minorticks_on()

plt.savefig('num_of_diff_orders.pdf', format='pdf', dpi=600, bbox_inches='tight')
fig.set_dpi(500)

plt.show()
```
