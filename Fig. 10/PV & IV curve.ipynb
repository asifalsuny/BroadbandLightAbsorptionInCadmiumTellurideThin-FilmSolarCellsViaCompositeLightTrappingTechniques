```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
from scipy import interpolate
from matplotlib.ticker import MultipleLocator
import h5py
import matplotlib.ticker as mtick
plt.rcParams["font.family"] = "Times New Roman"
plt.rc('font', weight='normal')
```

```python
mat_file = h5py.File('PIV_data.mat')

I = mat_file['J'][:]
V = mat_file['V'][:] 
P = mat_file['P'][:]

I = I.T
V = V.T
P = P.T

I = I[slice(None),0]
V = V[slice(None),0]
P = P[slice(None),0]

#--------------------------------------------#

I1 = mat_file['J1'][:]
V1 = mat_file['V1'][:] 
P1 = mat_file['P1'][:]

I1 = I1.T
V1 = V1.T
P1 = P1.T

I1 = I1[slice(None),0]
V1 = V1[slice(None),0]
P1 = P1[slice(None),0]

#--------------------------------------------#

I2 = mat_file['J2'][:]
V2 = mat_file['V2'][:] 
P2 = mat_file['P2'][:]

I2 = I2.T
V2 = V2.T
P2 = P2.T

I2 = I2[slice(None),0]
V2 = V2[slice(None),0]
P2 = P2[slice(None),0]



fig = plt.figure()
fig, ax1 = plt.subplots(figsize=(4.28,3))
# axs.grid(True)

ax1.set_xlabel('Voltage (V)', weight='normal', fontsize=16)  # Set font size for x-axis label
ax1.set_ylabel('Current density, J (mA/cm$^2$)', color='black', weight='normal', fontsize=16)  # Set font size for y-axis label

# ax1.tick_params(axis='x', labelcolor='black', labelsize=5)  # Set font size for tick labels on both axes
# ax1.tick_params(axis='y', labelcolor='black', labelsize=5)  # Set font size for tick labels on both axes

ax1.plot(V, I, 'o', color='red', linestyle='-', markersize=4, label='Bare CdTe TFSC')
ax1.plot(V1, I1, 'd', color='green', linestyle='-', markersize=4, label='CdTe TFSC with NC texture')
ax1.plot(V2, I2, 'h', color='blue', linestyle='-', markersize=4, label='CdTe TFSC with NC texture & Ge NP')

ax1.set_xlim([V.min(), V.max()+.04])
ax1.set_ylim([0, 37])

# Set the y-axis major locator with a step of 5 units
ax1.yaxis.set_major_locator(MultipleLocator(5))

# Add legends
ax1.legend(loc='lower left', fontsize=9)

# Customize tick marks (inside, bigger size, and thicker)
# axs.tick_params(direction='in', length=6, width=1.5, which='both', axis='x')
ax1.tick_params(direction='in', which='both', labelsize=14)
ax1.minorticks_on()
# from matplotlib.ticker import AutoMinorLocator
# axs.yaxis.set_minor_locator(AutoMinorLocator())  # Enables minor ticks only for y-axis

plt.savefig('IV.pdf', format='pdf', dpi=600, bbox_inches='tight')

fig.set_dpi(500)
```
